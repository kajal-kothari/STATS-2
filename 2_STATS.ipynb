3. What is the significance level in hypothesis testing, and why is it important?

The significance level, denoted by α (alpha), is the probability of rejecting the null hypothesis when it is actually true (i.e., making a Type I error). It is a predetermined threshold, typically set at 0.05 (5%) or 0.01 (1%), which defines how much risk of a Type I error a researcher is willing to accept. Its importance lies in providing a benchmark for decision-making: if the P-value is less than α, the result is considered statistically significant, leading to the rejection of H

4. What does a P-value represent in hypothesis testing?

A P-value (probability value) in hypothesis testing represents the probability of obtaining test results as extreme as, or more extreme than, the observed results, assuming that the null hypothesis is true. In simpler terms, it quantifies the strength of the evidence against the null hypothesis. A small P-value indicates that the observed data would be very unlikely if the null hypothesis were true, thereby providing strong evidence against H

5. How do you interpret the P-value in hypothesis testing?

Interpreting the P-value involves comparing it to the predetermined significance level (α). If the P-value is less than or equal to α (P≤α), it indicates that the observed data are statistically significant, providing sufficient evidence to reject the null hypothesis. This means the observed effect is unlikely to have occurred by random chance alone. If the P-value is greater than α (P>α), then the results are not statistically significant, and there is insufficient evidence to reject the null hypothesis. This does not mean the null hypothesis is true, but rather that we lack enough evidence to dispute it.

8. What is the Z-test, and when is it used in hypothesis testing?

The Z-test is a parametric statistical test used to determine if there is a significant difference between a sample mean and a population mean, or between two sample means, when the population standard deviation is known. It is used when comparing means, testing proportions, or constructing confidence intervals for population means or proportions, particularly when population variances are known.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

The T-distribution (Student's t-distribution) is a probability distribution similar in shape to the normal distribution but with heavier tails, meaning it has more probability in the tails and less in the center. It is characterized by its degrees of freedom (df). It should be used instead of the normal distribution in hypothesis testing when:

The population standard deviation is unknown.
The sample size is small (typically n<30). In such cases, using the T-distribution accounts for the additional uncertainty introduced by estimating the population standard deviation from the sample data, providing more accurate critical values and P-values. As the sample size increases, the T-distribution approaches the normal distribution.
11. What is the difference between a Z-test and a T-test?

The primary difference between a Z-test and a T-test lies in their assumptions regarding the population standard deviation and sample size:

Z-test: Used when the population standard deviation (σ) is known, or when the sample size (n) is large (typically n≥30), allowing the use of the Central Limit Theorem. It relies on the normal distribution.
T-test: Used when the population standard deviation (σ) is unknown and must be estimated from the sample standard deviation (s), especially when the sample size (n) is small (typically n<30). It relies on the T-distribution, which accounts for the increased uncertainty due to the unknown population standard deviation. In essence, the T-test is a more conservative choice when population parameters are less certain.
12. What is the T-test, and how is it used in hypothesis testing?

The T-test is a parametric statistical test used to compare means. It is particularly useful when the population standard deviation is unknown and the sample size is relatively small. In hypothesis testing, it's used to:

Compare a sample mean to a hypothesized population mean (one-sample t-test).
Compare the means of two independent groups (independent samples t-test).
Compare the means of two related or dependent groups (paired samples t-test). The T-test calculates a t-statistic, which is then compared to critical values from the T-distribution (based on degrees of freedom) or used to calculate a P-value to determine whether to reject the null hypothesis of no difference between means.
13. What is the relationship between Z-test and T-test in hypothesis testing?

The Z-test and T-test are both parametric tests used for hypothesis testing about population means. Their relationship is one of specificity and generalization. The Z-test is applicable under more restrictive conditions (known population standard deviation or very large sample size). The T-test, on the other hand, is a more robust and frequently used alternative when the population standard deviation is unknown, which is a common scenario in real-world research. As the sample size increases, the T-distribution approaches the normal distribution, and consequently, the results of a T-test will approximate those of a Z-test. Thus, the T-test can be seen as a more general form of the Z-test, accommodating more practical scenarios.

14. What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain an unknown population parameter (e.g., mean, proportion) with a certain level of confidence. It is expressed as a percentage (e.g., 95% CI). For interpreting statistical results, if the confidence interval for a difference between groups does not include zero, it suggests a statistically significant difference. Similarly, if the confidence interval for a parameter does not include a hypothesized value from the null hypothesis, it indicates that the null hypothesis can be rejected. It provides both an estimate of the parameter and a measure of the precision and uncertainty of that estimate.

15. What is the margin of error, and how does it affect the confidence interval?

The margin of error (MOE) quantifies the amount of random sampling error in a survey or statistical result. It is the radius (or half-width) of a confidence interval. It is calculated based on the chosen confidence level, the variability of the data (standard deviation/error), and the sample size. The margin of error directly affects the width of the confidence interval:

Larger margin of error: Results in a wider confidence interval, indicating less precision in the estimate of the population parameter. This can be due to a smaller sample size, higher variability, or a higher confidence level.
Smaller margin of error: Results in a narrower confidence interval, indicating greater precision. This is desirable and is typically achieved with a larger sample size, lower variability, or a lower confidence level.

19. What is the F-distribution, and when is it used in hypothesis testing?

The F-distribution (Fisher-Snedecor distribution) is a right-skewed probability distribution that arises from the ratio of two independent chi-square random variables, each divided by its degrees of freedom. It is characterized by two sets of degrees of freedom: numerator and denominator degrees of freedom. It is primarily used in hypothesis testing for:

ANOVA (Analysis of Variance) tests: To compare the means of three or more groups simultaneously by analyzing the variance between and within groups.
Testing the equality of two population variances: To determine if two population variances are significantly different. The F-statistic is the cornerstone of ANOVA, providing a means to assess if observed differences in group means are statistically significant or merely due to random chance.
20. What is an ANOVA test, and what are its assumptions?

ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more independent groups to determine if at least one group mean is significantly different from the others. It does this by partitioning the total variance in the data into variance attributable to group differences (between-group variance) and variance attributable to random error (within-group variance).
Its key assumptions are:

Independence of observations: Data points within and between groups must be independent.
Normality: The dependent variable should be approximately normally distributed within each group.
Homoscedasticity (Homogeneity of variances): The variance of the dependent variable should be approximately equal across all groups. Violations of these assumptions can affect the validity of the ANOVA results.
21. What are the different types of ANOVA tests?

There are several types of ANOVA tests, each suited for different research designs:

One-Way ANOVA: Used when comparing the means of three or more independent groups based on one categorical independent variable (factor).
Two-Way ANOVA: Used when comparing means across two independent categorical variables (factors) and also to examine their interaction effect on a dependent variable.
MANOVA (Multivariate ANOVA): Used when there are two or more dependent variables that are related, and one or more independent categorical variables.
Repeated Measures ANOVA: Used when the same subjects are measured multiple times under different conditions or at different time points (i.e., dependent groups). Each type extends the basic ANOVA principle to handle more complex experimental designs.
22. What is the F-test, and how does it relate to hypothesis testing?

The F-test is a statistical test based on the F-distribution, primarily used to compare variances or to compare means in designs with three or more groups. In hypothesis testing, it calculates an F-statistic, which is the ratio of two variances.
Its relation to hypothesis testing is crucial:

ANOVA: The F-test is the core of ANOVA. It tests the null hypothesis that all group means are equal. The F-statistic compares the variance between groups to the variance within groups. A large F-statistic suggests that the variation between group means is greater than what would be expected by chance, leading to the rejection of the null hypothesis and indicating at least one group mean is different.
Comparing Variances: It can also be used to test the null hypothesis that two population variances are equal, a common assumption check for other tests like the independent samples T-test.

```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results
import numpy as np
from scipy.stats import norm

def z_test_one_sample(sample_data, pop_mean, pop_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    std_err = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / std_err
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print("✅ Result: Reject the null hypothesis (significant difference).")
    else:
        print("❌ Result: Fail to reject the null hypothesis (no significant difference).")

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=52, scale=10, size=30)
z_test_one_sample(sample_data=sample, pop_mean=50, pop_std=10)
```

```python
#2 Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python@
import numpy as np
from scipy.stats import ttest_1samp

def simulate_and_hypothesis_test(mu=50, sample_size=30, alpha=0.05):
    data = np.random.normal(loc=mu + 2, scale=10, size=sample_size)  # shifted mean to create effect
    t_stat, p_value = ttest_1samp(data, popmean=mu)

    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < alpha:
        print("✅ Reject the null hypothesis.")
    else:
        print("❌ Fail to reject the null hypothesis.")

# Run it
simulate_and_hypothesis_test()
```

```python
#3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean@
from scipy.stats import norm

def one_sample_ztest(data, mu, sigma, alpha=0.05):
    mean = np.mean(data)
    n = len(data)
    z = (mean - mu) / (sigma / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))

    print(f"Z = {z:.4f}, P = {p:.4f}")
    if p < alpha:
        print("✅ Reject H₀: Sample mean is significantly different.")
    else:
        print("❌ Fail to reject H₀.")

# Example
data = np.random.normal(52, 10, 40)
one_sample_ztest(data, mu=50, sigma=10)
```

```python
#4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot@
import matplotlib.pyplot as plt

def visualize_z_test(z_stat, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    z_crit = norm.ppf(1 - alpha / 2)
    plt.plot(x, y, label='Standard Normal Curve')
    plt.axvline(z_stat, color='red', linestyle='--', label=f'Z = {z_stat:.2f}')
    plt.axvline(z_crit, color='green', linestyle=':', label=f'Z critical = ±{z_crit:.2f}')
    plt.axvline(-z_crit, color='green', linestyle=':')
    plt.fill_between(x, y, where=(x < -z_crit) | (x > z_crit), color='gray', alpha=0.3, label='Rejection Region')

    plt.title("Two-tailed Z-test Decision Region")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example
visualize_z_test(z_stat=2.1)
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing@
def plot_type_errors(mu0=0, mu1=1, sigma=1, alpha=0.05):
    x = np.linspace(-4, 5, 1000)
    y0 = norm.pdf(x, mu0, sigma)
    y1 = norm.pdf(x, mu1, sigma)

    z_crit = norm.ppf(1 - alpha)
    plt.plot(x, y0, label='H₀ True', color='blue')
    plt.plot(x, y1, label='H₁ True', color='red')

    plt.fill_between(x, y0, where=(x > z_crit), color='blue', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y1, where=(x <= z_crit), color='red', alpha=0.3, label='Type II Error')

    plt.axvline(z_crit, linestyle='--', color='black', label=f'Z critical = {z_crit:.2f}')
    plt.title("Type I and Type II Errors")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run it
plot_type_errors()
```

```python
#6 Write a Python program to perform an independent T-test and interpret the results@
from scipy.stats import ttest_ind

def independent_ttest():
    group1 = np.random.normal(60, 10, 30)
    group2 = np.random.normal(65, 10, 30)

    t_stat, p = ttest_ind(group1, group2)

    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p:.4f}")
    if p < 0.05:
        print("✅ Groups are significantly different.")
    else:
        print("❌ No significant difference between groups.")

# Run it
independent_ttest()
```

```python
#7  Perform a paired sample T-test using Python and visualize the comparison results@
from scipy.stats import ttest_rel

def paired_ttest():
    before = np.random.normal(70, 10, 30)
    after = before + np.random.normal(2, 3, 30)  # small improvement

    t_stat, p = ttest_rel(before, after)

    print(f"Paired T-test P-value: {p:.4f}")
    plt.plot(before, label='Before')
    plt.plot(after, label='After')
    plt.title("Paired Sample Comparison")
    plt.legend()
    plt.show()

# Run it
paired_ttest()
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(data, confidence=0.95):
    import scipy.stats as st

    n = len(data)
    mean = np.mean(data)
    stderr = st.sem(data)
    margin = st.t.ppf((1 + confidence) / 2, n - 1) * stderr
    ci = (mean - margin, mean + margin)

    print(f"{int(confidence*100)}% CI: ({ci[0]:.2f}, {ci[1]:.2f})")
    print("Meaning: We're", int(confidence*100), "% confident that the true population mean lies within this interval.")

# Example usage
data = np.random.normal(100, 15, 40)
confidence_interval(data)
```

```python
#10  Write a Python program to calculate the margin of error for a given confidence level using sample dataD
import numpy as np
from scipy.stats import norm

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_dev = np.std(data, ddof=1)
    z = norm.ppf(1 - (1 - confidence)/2)
    moe = z * (std_dev / np.sqrt(n))

    print(f"Margin of Error ({int(confidence*100)}% confidence): ±{moe:.2f}")

# Example
data = np.random.normal(100, 15, 50)
margin_of_error(data)
```

```python
#11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the processD
def bayes_theorem(p_A, p_B_given_A, p_B_given_not_A):
    p_not_A = 1 - p_A
    p_B = p_B_given_A * p_A + p_B_given_not_A * p_not_A
    p_A_given_B = (p_B_given_A * p_A) / p_B

    print(f"P(A|B) = {p_A_given_B:.4f}")

# Example
bayes_theorem(p_A=0.01, p_B_given_A=0.95, p_B_given_not_A=0.05)
```

```python
#17  Write a Python program to perform an ANOVA test to compare means between multiple groups and
#interpret the results
from scipy.stats import f_oneway

def one_way_anova():
    g1 = np.random.normal(70, 10, 30)
    g2 = np.random.normal(75, 10, 30)
    g3 = np.random.normal(80, 10, 30)

    f_stat, p = f_oneway(g1, g2, g3)
    print(f"F = {f_stat:.2f}, p = {p:.4f}")
    if p < 0.05:
        print("✅ At least one group mean is significantly different.")
    else:
        print("❌ No significant difference between group means.")

# Run it
one_way_anova()
```

```python
#19  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVAD
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    for i, g in enumerate(groups):
        _, p = shapiro(g)
        print(f"Group {i+1} Normality p-value: {p:.4f}")

    _, p_levene = levene(*groups)
    print(f"Levene’s test (Equal variance) p-value: {p_levene:.4f}")

# Example usage
g1 = np.random.normal(60, 10, 30)
g2 = np.random.normal(70, 10, 30)
g3 = np.random.normal(65, 10, 30)
check_anova_assumptions(g1, g2, g3)
```

```python
#22 Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group meansD
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

def one_way_anova_with_boxplot():
    # Simulate data for three groups
    np.random.seed(42)
    group_a = np.random.normal(70, 10, 30)
    group_b = np.random.normal(75, 10, 30)
    group_c = np.random.normal(80, 10, 30)

    # Combine into a DataFrame
    df = pd.DataFrame({
        'score': np.concatenate([group_a, group_b, group_c]),
        'group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30
    })

    # Perform one-way ANOVA
    f_stat, p_value = f_oneway(group_a, group_b, group_c)
    print("📊 One-Way ANOVA Results")
    print(f"F-statistic = {f_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < 0.05:
        print("✅ Conclusion: At least one group mean is significantly different.")
    else:
        print("❌ Conclusion: No significant difference between group means.")

    # Plot boxplots for visualization
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='group', y='score', data=df, palette='Set2')
    plt.title('Group Comparison Using One-Way ANOVA')
    plt.ylabel('Score')
    plt.grid(True)
    plt.show()

# Run the function
one_way_anova_with_boxplot()
```

```python
#23 7"D Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the meansD
import numpy as np
from scipy.stats import ttest_ind

def simulate_and_test_means():
    np.random.seed(42)

    # Simulate two groups with slightly different means
    group1 = np.random.normal(loc=100, scale=15, size=50)
    group2 = np.random.normal(loc=105, scale=15, size=50)

    # Perform independent two-sample T-test
    t_stat, p_value = ttest_ind(group1, group2)

    print("📊 Independent Two-Sample T-Test")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < 0.05:
        print("✅ Conclusion: There is a significant difference between the group means.")
    else:
        print("❌ Conclusion: No significant difference between the group means.")

# Run the function
simulate_and_test_means()
```

```python
#24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the resultsD
import numpy as np
from scipy.stats import chi2

def variance_test_chi_square(sample, sigma_squared_0, alpha=0.05):
    n = len(sample)
    sample_variance = np.var(sample, ddof=1)
    chi2_stat = (n - 1) * sample_variance / sigma_squared_0

    # Two-tailed test
    chi2_critical_low = chi2.ppf(alpha / 2, df=n - 1)
    chi2_critical_high = chi2.ppf(1 - alpha / 2, df=n - 1)

    print("📊 Chi-Square Test for Population Variance")
    print(f"Sample Variance = {sample_variance:.2f}")
    print(f"Chi2 Statistic = {chi2_stat:.2f}")
    print(f"Critical Region: ({chi2_critical_low:.2f}, {chi2_critical_high:.2f})")

    if chi2_stat < chi2_critical_low or chi2_stat > chi2_critical_high:
        print("✅ Conclusion: Reject H₀ — the sample variance differs significantly.")
    else:
        print("❌ Conclusion: Fail to reject H₀ — no significant difference in variance.")

# Example
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=12, size=30)
variance_test_chi_square(sample=sample_data, sigma_squared_0=144)  # 144 = 12²
```

```python
#25  Write a Python script to perform a Z-test for comparing proportions between two datasets or groupsD
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(successes1, size1, successes2, size2, alpha=0.05):
    count = np.array([successes1, successes2])
    nobs = np.array([size1, size2])
    z_stat, p_value = proportions_ztest(count, nobs)

    print("📊 Two-Proportion Z-Test")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print("✅ Conclusion: Proportions are significantly different.")
    else:
        print("❌ Conclusion: No significant difference between proportions.")

# Example
# Group 1: 45 successes out of 100
# Group 2: 30 successes out of 100
z_test_proportions(45, 100, 30, 100)
```

```python
#26  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the resultsD
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def f_test_and_visualize(data1, data2, alpha=0.05):
    # Calculate sample variances
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)

    # Ensure larger variance is in numerator (to keep F > 1)
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = len(data1) - 1, len(data2) - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = len(data2) - 1, len(data1) - 1

    # Critical values
    f_critical = f.ppf(1 - alpha, dfn, dfd)

    print("📊 F-Test for Comparing Variances")
    print(f"F-statistic = {f_stat:.4f}")
    print(f"Critical value (α={alpha}): {f_critical:.4f}")

    if f_stat > f_critical:
        print("✅ Conclusion: Reject H₀ — Variances are significantly different.")
    else:
        print("❌ Conclusion: Fail to reject H₀ — No significant difference in variances.")

    # Visualization
    x = np.linspace(0.1, 5, 1000)
    y = f.pdf(x, dfn, dfd)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'F-distribution (dfn={dfn}, dfd={dfd})')
    plt.axvline(f_stat, color='red', linestyle='--', label=f'F = {f_stat:.2f}')
    plt.axvline(f_critical, color='green', linestyle='--', label=f'Critical F = {f_critical:.2f}')
    plt.fill_between(x, y, where=(x >= f_critical), color='green', alpha=0.3, label='Rejection Region')
    plt.title("F-Test: Comparing Two Variances")
    plt.xlabel("F value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Simulate example data
np.random.seed(0)
data1 = np.random.normal(50, 5, 30)
data2 = np.random.normal(50, 10, 30)

f_test_and_visualize(data1, data2)
```

```python
#27  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

def chi_square_gof_simulated():
    # Simulate observed frequencies (e.g., from rolling a die)
    observed = np.random.randint(15, 30, size=6)

    # Expected: uniform distribution (fair die)
    expected = [np.mean(observed)] * 6

    # Perform the Chi-square goodness-of-fit test
    chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

    print("🎲 Chi-Square Goodness-of-Fit Test")
    print(f"Observed: {observed}")
    print(f"Expected: {expected}")
    print(f"Chi2 = {chi2_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < 0.05:
        print("✅ Conclusion: Reject H₀ — Observed data does not fit the expected distribution.")
    else:
        print("❌ Conclusion: Fail to reject H₀ — Data fits the expected distribution.")

    # Visualization
    categories = [f'Cat {i+1}' for i in range(len(observed))]
    x = np.arange(len(categories))

    plt.bar(x - 0.2, observed, width=0.4, label='Observed', color='skyblue')
    plt.bar(x + 0.2, expected, width=0.4, label='Expected', color='orange')
    plt.xticks(x, categories)
    plt.title("Observed vs Expected Frequencies")
    plt.ylabel("Count")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run it
chi_square_gof_simulated()
```
